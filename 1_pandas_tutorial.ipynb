## Series
- 1차원 배열
- label을 정해줄 수 있음
    - numpy.array와 차이점!

```python
exam = [90, 70, 100, 95, 85]
examseries = pd.Series(exam)
examseries
```

```python
# Series labeling
examseries = pd.Series(exam, 
                       index = ['korean', 'english', 'math', 'society', 'science'])
examseries
```

```python
examseries['math']
```

```python
# create Series using Dictionary
dic_exam = {'korean':95, 'english':70, 'math':100, 'society':95, 'science':85}
examseries = pd.Series(dic_exam)
examseries
```

## Cleaning Data
- 비어있는 경우
- 데이터 포맷이 잘못된 경우
- 데이터가 잘못된경우
- 데이터 중복

```python
df = pd.read_csv('./dataset/odd_data.txt', sep=',', index_col=0)
df
```

### 비어있는 경우 삭제
- df.dropna() : 비어있는 셀이 있는 행 삭제
    - 파라미터 `inplace = True` : 새로운 객체를 반환하지 않고, 원본 객체를 수정

```python
new_df = df.dropna() # 또는 df.dropna(inplace = True)
new_df
```

### 비어있는 셀 채우기
- df.fillna(value, `inplace = True`) : df의 비어있는 셀을 value로 채움
    - df.col_name.mean() : 평균
    - df.col_name.median() : 중앙값
    - df.col_name.mode()[0] : 최빈값
        - 최빈값이 여러개인 경우도 있을 수 있어서 mode()는 Series type으로 return함

```python
df = pd.read_csv('./dataset/odd_data.txt', sep=',', index_col=0)
df.fillna(df.Duration.mean(), inplace = True)
print(df.to_string)
```

### 잘못된 타입인 경우
- to_datetime : 날짜 타입으로 변환
    - 날짜가 아닌, 완전히 다른 타입의 데이터는 NaT(Not a Time)로 바꿈(=Nan)

```python
df = pd.read_csv('./dataset/odd_data.txt', sep=',', index_col=0)
df
```

```python
df.Date = pd.to_datetime(df.Date)
df
```

### 잘못된 데이터 선택해서 고치기
- df.loc[row_index, col_name] : 특정 셀 선택해서 변경
- df.drop(row_index, inplace=True) : 잘못된 셀 있는 행 자체를 삭제

```python
df = pd.read_csv('./dataset/odd_data.txt', sep=',', index_col=0)
df
```

```python
df.loc[7, 'Duration'] = 45
df
```

### 중복된 행
- df.duplicated() : 중복된 행 체크
- df.drop_duplicates(inplace = True) : 중복된 행 삭제

```python
# 11행과 12행이 중복됨
df = pd.read_csv('./dataset/odd_data.txt', sep=',', index_col=0)
df
```

```python
df.duplicated()
```

## Plot

    - type == *scatter* : scatter plot
